```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
import numpy as np

def load_iris_dataset(path):
    """Load the Iris dataset from a CSV file."""
    data = np.genfromtxt(path, delimiter=',', skip_header=1)
    X = data[:, :-1]  # Features
    y = data[:, -1]   # Target
    return X, y

# Path to the Iris dataset file
path_to_dataset = 'D:\\4.2 work\\ML\\Irislec.csv'  # Corrected file path

# Load the Iris dataset
X_train, y_train = load_iris_dataset(path_to_dataset)

# Print the loaded dataset
print("Features (X):")
print(X_train)
print("Target (y):")
print(y_train)
```

```python
def euclidean_distance(point1, point2):
    """Calculate the Euclidean distance between two points."""
    return np.sqrt(np.sum((point1 - point2)**2))

# Loading the dataset from the CSV file
path_to_dataset = 'D:\\4.2 work\\ML\\Irislec.csv'
data = np.genfromtxt(path_to_dataset, delimiter=',', skip_header=1)

# Separating features (X) and target (y)
X = data[:, 1:-1]  # Features (excluding the first column which contains index and the last column which contains the target)
y = data[:, -1]    # Target

# New observation with the same number of features as the data points in X
new_observation = np.array([5.2, 3.1, 1.4, 0.2])  

# Computing Euclidean distances between the new observation and each data point in X
distances = [euclidean_distance(new_observation, x) for x in X]

# Printing the Euclidean distances
print("Euclidean distances between the new observation and each data point:")
for idx, distance in enumerate(distances, start=1):
    print(f"Distance to data point {idx}: {distance:.4f}")
```

```python
# Ranking distances from nearest to furthest
sorted_indices = np.argsort(distances)

# Printing the ranked distances
print("Ranked distances from nearest to furthest:")
for rank, idx in enumerate(sorted_indices, start=1):
    print(f"Rank {rank}: Distance to data point {idx + 1}: {distances[idx]:.4f}")
```

```python
def k_nearest_neighbors(X, y, new_observation, k=5):
    """Classify a new observation using k-nearest neighbors."""
    # Compute distances between the new observation and all data points
    distances = [euclidean_distance(new_observation, x) for x in X]
    
    # Getting indices of k nearest neighbors
    nearest_indices = np.argsort(distances)[:k]
    
    # Getting classes of k nearest neighbors
    nearest_classes = y[nearest_indices]
    
    # Checking if there are any missing values in the nearest classes
    if np.isnan(nearest_classes).any():
        # Handle missing values, e.g., return a default class
        predicted_class = "Default"
    else:
        # Finding the most common class among the nearest neighbors
        unique_classes, class_counts = np.unique(nearest_classes, return_counts=True)
        predicted_class = unique_classes[np.argmax(class_counts)]
    
    return predicted_class

# Classify the new observation using k-nearest neighbors with k=5
predicted_class = k_nearest_neighbors(X, y, new_observation, k=5)

print("Predicted class for the new observation:", predicted_class)
```
